```python
data = []
box_data = []  # This will store the [time, distance] pairs

# Variables to compute averages
total_x, total_y, total_z, total_distance = 0, 0, 0, 0
count = 0

try:
    with open(path, 'r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip the header row
        for row in csv_reader:
            if len(row) >= 8:  # Ensure there are enough columns
                try:
                    time_box = float(row[1].replace("'", "").replace(",", "").strip())
                    x_position_box = float(row[-3].replace("'", "").replace(",", "").strip())
                    y_position_box = float(row[-2].replace("'", "").replace(",", "").strip())
                    z_position_box = float(row[-1].replace("'", "").replace(",", "").strip())
                    distance_box = np.sqrt(x_position_box**2 + y_position_box**2 + z_position_box**2)

                    # Accumulate sums for averaging
                    total_x += x_position_box
                    total_y += y_position_box
                    total_z += z_position_box
                    total_distance += distance_box
                    count += 1

                    data.append([time_box, x_position_box, y_position_box, z_position_box])
                    box_data.append([time_box, distance_box])  # Append [time, distance] to box_data
                except ValueError:
                    # Skip rows where conversion to float fails
                    continue
except FileNotFoundError:
    print("The file was not found. Check the file path and try again.")
except Exception as e:
    print(f"An error occurred: {e}")

# Print the cleaned and converted data
for item in data:
    print(item)

# Print the time and distance data
print("\nTime and Distance Data:")
for item in box_data:
    print(item)

# Calculate averages if count is not zero
avg_x = total_x / count if count > 0 else 0
avg_y = total_y / count if count > 0 else 0
avg_z = total_z / count if count > 0 else 0
avg_distance = total_distance / count if count > 0 else 0

# Print average values
print("\nAverage Values:")
print(f"Average X: {avg_x}")
print(f"Average Y: {avg_y}")
print(f"Average Z: {avg_z}")
print(f"Average Distance: {avg_distance}")

# Plot the data if it was successfully collected
if box_data:
    times = [d[0] for d in box_data]
    distances = [d[1] for d in box_data]

    plt.figure(figsize=(10, 5))
    plt.plot(times, distances, marker='o', linestyle='-', color='b')
    plt.title('Distance from Origin Over Time')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Distance (sqrt(x^2 + y^2 + z^2))')
    plt.grid(True)
    plt.show()
```

```python
# Define paths to CSV files
paths = {
    'large_angle': '/home/tiantan/Documents/Apriltag_distance_detection/data/Apriltag_video/Apriltag_video/large angle.csv',
    'medium_angle': '/home/tiantan/Documents/Apriltag_distance_detection/data/Apriltag_video/Apriltag_video/medium angle.csv',
    'small_angle': '/home/tiantan/Documents/Apriltag_distance_detection/data/Apriltag_video/Apriltag_video/small angle.csv',
    'straight_angle': '/home/tiantan/Documents/Apriltag_distance_detection/data/Apriltag_video/Apriltag_video/Straight.csv',
    'wide_len': '/home/tiantan/Documents/Apriltag_distance_detection/data/Apriltag_video/Apriltag_video/wide len.csv'
}

# Average coordinates from which to calculate distance
avg_x = -0.4409805301204824
avg_y = -0.17350980722891574
avg_z = 0.098411465060241

# Dictionary to store data
data_dict = {key: [] for key in paths}
distance_derivative_dict = {key: [] for key in paths}

def process_csv(path, label):
    data = []
    time_data = []
    distance_data = []
    
    try:
        with open(path, 'r') as file:
            csv_reader = csv.reader(file)
            next(csv_reader)  # Skip the header row
            for row in csv_reader:
                if len(row) >= 8:
                    try:
                        time = float(row[1].replace("'", "").replace(",", "").strip())
                        x = float(row[-3].replace("'", "").replace(",", "").strip())
                        y = float(row[-2].replace("'", "").replace(",", "").strip())
                        z = float(row[-1].replace("'", "").replace(",", "").strip())

                        # Calculate Euclidean distance from the average point
                        distance = np.sqrt((x - avg_x)**2 + (y - avg_y)**2 + (z - avg_z)**2)
                        
                        data.append([time, x, y, z, distance])
                        time_data.append(time)
                        distance_data.append(distance)
                    except ValueError:
                        continue
        data_dict[label] = data
        
        # Calculate the derivative of the distance with respect to time
        derivative = np.diff(distance_data) / np.diff(time_data)
        distance_derivative_dict[label] = derivative
        
        # Plot distance and its derivative
        plt.figure(figsize=(12, 6))
        plt.subplot(121)
        plt.plot(time_data, distance_data, label=f'Distance - {label}')
        plt.xlabel('Time (seconds)')
        plt.ylabel('Distance')
        plt.title('Distance over Time')
        plt.legend()
        plt.grid(True)
        
        plt.subplot(122)
        plt.plot(time_data[:-1], derivative, label=f'Derivative - {label}', color='red')
        plt.xlabel('Time (seconds)')
        plt.ylabel('Distance Derivative')
        plt.title('Derivative of Distance over Time')
        plt.legend()
        plt.grid(True)
        
        plt.tight_layout()
        plt.show()
        
        # Alert if there are significant changes in the derivative
        significant_changes = np.where(derivative > np.percentile(derivative, 95))[0]
        if significant_changes.size > 0:
            print(f"Notice: Significant increase in distance derivative for {label} at times {time_data[:-1][significant_changes]}")

    except FileNotFoundError:
        print(f"The file {path} was not found. Check the file path and try again.")
    except Exception as e:
        print(f"An error occurred while processing {path}: {e}")

# Process each CSV file
for label, path in paths.items():
    process_csv(path, label)
```
